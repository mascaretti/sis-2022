# Plots of MCMC

```python
import arviz as az
import numpy as np
import pickle
import matplotlib.pyplot as plt
```

```python
# ArviZ ships with style sheets!
az.style.use("arviz-darkgrid")
```

```python
file = "mcmc-2022-02-28-00-13-08.pkl"
file_name = f"../data/interim/{file}"
with (open(file_name, "rb")) as f:
    mcmc = pickle.load(f)
```

```python
np.round(np.mean(mcmc.get("Beta"), axis=0), 3)
```

```python
burn_in = 300
```

```python
mcmc.keys()
```

```python
def print_beta(i: int, j: int, burn_in) -> None:
    assert i >= 0 and i <= 2, "Betas have three rows"
    assert j >= 0 and j <= 1, "Betas have two columns"
    az.plot_trace(mcmc.get("Beta")[burn_in:, i, j])
```

```python
print_beta(0, 0, burn_in)
```

## $\mu$

```python
mu_0 = dict()
mu_0["mu_0"] = mcmc.get("mu")[burn_in:, 0]
az.plot_posterior(mu_0)
# plt.savefig("../reports/figures/weak_mu_0.pdf", format="pdf", bbox_inches="tight")
az.summary(mu_0)
```
